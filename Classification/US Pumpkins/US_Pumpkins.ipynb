```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load the dataset and print the first 5 rows
data = pd.read_csv('US-pumpkins.csv')
data.head()
```

```python
data = data[data['Package'].str.contains('bushel', case=True, regex=True)]
```

```python
from sklearn.preprocessing import LabelEncoder

new_columns = ['Color', 'Origin', 'Item Size', 'Variety', 'City Name', 'Package']
new_data = data.drop([i for i in data.columns if i not in new_columns], axis=1)
new_data.dropna(inplace=True)

new_data = new_data.apply(LabelEncoder().fit_transform)

new_data.info()
```

```python
g = sns.PairGrid(new_data)
g.map(sns.scatterplot)
```

```python
sns.swarmplot(x='Color', y='Item Size', data=new_data)
```

```python
sns.catplot(x='Color', y='Item Size', kind='violin', data=new_data)
```

```python
from sklearn.model_selection import train_test_split

selected_features = ['Origin', 'Item Size', 'Variety', 'City Name', 'Package']

X = new_data[selected_features]
y = new_data['Color']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

print(classification_report(y_test, predictions))
print(f"Predicted labels: {predictions}")
print(f"Accuracy: {accuracy_score(y_test, predictions)}")
```

```python
confusion_matrix(y_test, predictions)
```

```python
y_scores = model.predict_proba(X_test)
# calculatr ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores[:, 1])
sns.lineplot([0, 1], [0, 1])
sns.lineplot(fpr, tpr)
```

```python
auc = roc_auc_score(y_test, y_scores[:, 1])
auc
```
